```python
import pandas as pd
import json
import os
import pprint
```

```python
path = "C:\\Users\\Public\\phonepe\\pulse\\data\\aggregated\\transaction\\country\\india\\state\\"
Agg_state_list=os.listdir(path)
at = {'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [], 'Transaction_amount': []}

for state in Agg_state_list:
    state_path = os.path.join(path, state)

    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)

        for quarter_file in os.listdir(year_path):
            quarter_path = os.path.join(year_path, quarter_file)
            with open(quarter_path, 'r') as tran_file:
                c = json.load(tran_file)

                for i in c ['data']['transactionData']:
                    name = i['name']
                    count = i['paymentInstruments'][0]['count']
                    amount = i['paymentInstruments'][0]['amount']

                    at['Transaction_type'].append(name)
                    at['Transaction_count'].append(count)
                    at['Transaction_amount'].append(amount)
                    at['State'].append(state)
                    at['Year'].append(year)
                    at['Quarter'].append(int(quarter_file.strip('.json')))

#DataFrame
Agg_Trans = pd.DataFrame(at)
Agg_Trans.dtypes
```

```python
path = path = "C:\\Users\\Public\\phonepe\\pulse\\data\\aggregated\\user\\country\\india\\state\\"
Agg_user_list=os.listdir(path)
au = {'State': [], 'Year': [], 'Quarter': [], 'Device': [], 'User_count': [], 'Device_share_percent': []}

for state in Agg_user_list:
    state_path = os.path.join(path, state)

    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)

        for quarter_file in os.listdir(year_path):
            quarter_path = os.path.join(year_path, quarter_file)
            with open(quarter_path, 'r') as user_file:
                d = json.load(user_file)
                try:
                    for i in d['data']['usersByDevice']:
                        brand = i['brand']
                        count = i['count']
                        percentage = i['percentage']

                        au['Device'].append(brand)
                        au['User_count'].append(count)
                        au['Device_share_percent'].append(percentage)
                        au['State'].append(state)
                        au['Year'].append(year)
                        au['Quarter'].append(int(quarter_file.strip('.json')))
                except:
                    pass

Agg_user = pd.DataFrame(au)
Agg_user
```

```python
path="C:\\Users\\Public\\phonepe\\pulse\\data\\map\\transaction\\hover\\country\\india\\state\\"
Map_trans_list=os.listdir(path)

#dict to append the data
mt = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Transaction_type': [], 'Transaction_count': [], 'Total_amount': []}

for state in Map_trans_list:
    state_path = os.path.join(path, state)

    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)

        for quarter_file in os.listdir(year_path):
            quarter_path = os.path.join(year_path, quarter_file)

            with open(quarter_path, 'r') as data_file:
                b = json.load(data_file)

                for i in b['data']['hoverDataList']:
                    name = i['name']
                    type = i['metric'][0]['type']
                    count = i['metric'][0]['count']
                    amount = i['metric'][0]['amount']

                    mt['District'].append(name)
                    mt['Transaction_type'].append(type)
                    mt['Transaction_count'].append(count)
                    mt['Total_amount'].append(amount)
                    mt['State'].append(state)
                    mt['Year'].append(year)
                    mt['Quarter'].append(int(quarter_file.strip('.json')))

# Successfully created a DataFrame
Map_trans = pd.DataFrame(mt)
Map_trans
```

```python
path="C:\\Users\\Public\\phonepe\\pulse\\data\\map\\user\\hover\\country\\india\\state\\"
Map_user_list=os.listdir(path)

mu = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Reg_users': [], 'App_opens': []}

for state in Map_user_list:
    state_path = os.path.join(path, state)

    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)

        for quarter_file in os.listdir(year_path):
            quarter_path = os.path.join(year_path, quarter_file)
            with open(quarter_path, 'r') as data_file:
                h = json.load(data_file)

                for i in h ["data"]["hoverData"].items():
                    district = i[0]
                    registereduser = i[1]["registeredUsers"]
                    appOpens = i[1]['appOpens']

                    mu['District'].append(district)
                    mu['Reg_users'].append(registereduser)
                    mu['App_opens'].append(appOpens)
                    mu['State'].append(state)
                    mu['Year'].append(year)
                    mu['Quarter'].append(int(quarter_file.strip('.json')))

Map_user = pd.DataFrame(mu)
Map_user
```

```python
import pymysql
db = pymysql.connect(
        host="localhost",
        user="root",
        password="123",
        database="project_2")
        

mycursor = db.cursor()
mycursor.execute("use project_2")
```
